### mpg 통합 연비 변수 만들기

```python
# mpg 데이터 불러오기
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2  # 통합 연비 변수 만들기
mpg.head()
```

```python
mpg['mpg_mean']=mpg[['cty','hwy']].mean(axis='columns')
mpg.head()
```

```python
sum(mpg['total'])/len(mpg) # total 합계를 행 수로 나누기
```

```python
mpg['total'].mean() #통합 연비 변수 평균
```
